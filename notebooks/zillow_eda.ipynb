```python
import pathlib
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

root_path = pathlib.Path(".")
data_path = root_path / "data"
zillow_data_path = data_path / "zillow"

data = pd.read_csv(zillow_data_path / "Zip_zori_uc_sfrcondomfr_sm_month.csv").set_index("RegionName")
```

```python
# data.loc[11238]
time_series_cols = [col for col in data.columns if col.startswith("20")]

zip_codes = [11238, 11216, 11217, 11215, 11201]
fig, ax = plt.subplots()

for zip_code in zip_codes:
    ax.plot(data.loc[zip_code][time_series_cols], label=zip_code)
    
plt.legend()
```
